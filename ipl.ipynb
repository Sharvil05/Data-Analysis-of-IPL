```
The goal of this analysis is to gain insights from the Indian Premier League (IPL) dataset (2008-2023) by visualizing key trends, patterns, and statistics related to teams, players, matches, and venues. Using Python (Pandas, Matplotlib, Seaborn), we will explore the dataset to identify meaningful insights.
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/content/Ipl-clean-data2008-2023.csv"
df = pd.read_csv(file_path)

# Fix column names
df.columns = df.columns.str.strip().str.lower()

# Set font family to fix emoji issue
plt.rcParams['font.family'] = 'DejaVu Sans'

# Plot matches per season (Fixed FutureWarning & Emoji Issue)
plt.figure(figsize=(10, 5))
sns.countplot(x=df['season'], hue=df['season'], palette='coolwarm', legend=False)  # Fix for palette warning
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.title("Number of Matches Per Season (2008-2023)")  # Removed 📊 emoji
plt.xticks(rotation=45)
plt.show()
```

```
### 2️⃣ Toss Decision Analysis ###
plt.figure(figsize=(8, 6))
sns.countplot(x=df['decision'], hue=df['decision'], palette='coolwarm', legend=False)
plt.xlabel("Toss Decision")
plt.ylabel("Count")
plt.title("Toss Decision Distribution (Bat or Field)")
plt.show()
```

```
### 3️⃣ Matches Won Per Season ###
plt.figure(figsize=(12, 6))
sns.countplot(x=df['season'], hue=df['season'], order=df['season'].value_counts().index, palette="magma", legend=False)
plt.xlabel("Season")
plt.ylabel("Matches Won")
plt.title("Matches Won Per Season")
plt.xticks(rotation=45)
plt.show()
```

```
### 4️⃣ Most Man of the Match Awards ###
plt.figure(figsize=(12, 6))
df['pom'].value_counts().head(10).plot(kind='bar', color='lightcoral')
plt.xlabel("Players")
plt.ylabel("Total Awards")
plt.title("Top 10 Players with Most Man of the Match Awards")
plt.xticks(rotation=45)
plt.show()
```

```
### 5️⃣ Win Ratio in Super Over Matches (Fixed FutureWarning) ###
plt.figure(figsize=(8, 6))
df['super_over'] = df['super_over'].astype(str)  # Convert to string
sns.countplot(x=df['super_over'], hue=df['super_over'], palette="coolwarm", legend=False)
plt.xlabel("Super Over (Yes/No)")
plt.ylabel("Count")
plt.title("Number of Matches that Went to Super Over")
plt.show()
```

```
### 6️⃣ Home vs Away Wins ###
plt.figure(figsize=(12, 6))
df['home_win'] = df['winner'] == df['home_team']
sns.countplot(x=df['home_win'], hue=df['home_win'], palette="viridis", legend=False)
plt.xlabel("Win Type")
plt.ylabel("Count")
plt.title("Home Wins vs Away Wins")
plt.xticks([0, 1], ["Away Wins", "Home Wins"])
plt.show()
```

```
### 7️⃣ 📍 Matches Played at Different Venues ###
plt.figure(figsize=(14, 6))
df['venue_name'].value_counts().head(10).plot(kind='barh', color='orange')
plt.xlabel("Total Matches")
plt.ylabel("Venue")
plt.title("Top 10 Venues with Most IPL Matches (2008-2023)")
plt.gca().invert_yaxis()  # To display the highest bar at the top
plt.show()
```
